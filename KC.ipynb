```python
import pandas as pd
import vtk

# 读取Excel文件
df = pd.read_excel(r'D:software\lily-pad\sanjiaosanyuanzhu\KC=1.1.xlsx')

# 假设X, Y, Z列分别在第5, 6, 0列
X = df.iloc[:, 5].values
Y = df.iloc[:, 6].values
Z = df.iloc[:, 0].values

# 创建点集
points = vtk.vtkPoints()
for x, y, z in zip(X, Y, Z):
    points.InsertNextPoint(x, y, z)

# 创建PolyData
polyData = vtk.vtkPolyData()
polyData.SetPoints(points)

# 使用Delaunay3D算法来创建网格
delaunay = vtk.vtkDelaunay3D()
delaunay.SetInputData(polyData)

# 映射器和演员为原始网格
mapper_original = vtk.vtkPolyDataMapper()
mapper_original.SetInputConnection(delaunay.GetOutputPort())
actor_original = vtk.vtkActor()
actor_original.SetMapper(mapper_original)
actor_original.GetProperty().SetPointSize(50)  # 设置点的大小，以便区分
actor_original.GetProperty().SetOpacity(0.5)  # 设置透明度
# 由于Delaunay可能产生vtkUnstructuredGrid，我们使用DataSetSurfaceFilter来提取表面
surface_filter = vtk.vtkDataSetSurfaceFilter()
surface_filter.SetInputConnection(delaunay.GetOutputPort())


# 使用平滑滤波器来平滑网格
smoother = vtk.vtkSmoothPolyDataFilter()
smoother.SetInputConnection(surface_filter.GetOutputPort())
smoother.SetNumberOfIterations(50)

# 映射器
mapper = vtk.vtkPolyDataMapper()
mapper.SetInputConnection(smoother.GetOutputPort())

# 演员
actor = vtk.vtkActor()
actor.SetMapper(mapper)

# 创建渲染器、渲染窗口和交互式渲染器
renderer = vtk.vtkRenderer()
renderWindow = vtk.vtkRenderWindow()
renderWindow.AddRenderer(renderer)
renderWindowInteractor = vtk.vtkRenderWindowInteractor()
renderWindowInteractor.SetRenderWindow(renderWindow)

# 将演员添加到渲染器
renderer.AddActor(actor)
#renderer.AddActor(actor_original)
renderer.SetBackground(0.1, 0.2, 0.3)  # 设置背景颜色

# 初始化渲染窗口并开始交互
renderWindow.Render()
renderWindowInteractor.Start()
```

```python
import pandas as pd
import numpy as np
from scipy.optimize import curve_fit
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error, mean_absolute_error
# 读取Excel文件
df = pd.read_excel(r'D:software\lily-pad\sanjiaosanyuanzhu\KC=3.1.xlsx')
popt0 = [4, 0, 0, 0,1,0.5]  # 初始参数猜测

# 定义非线性模型函数
def nonlinear_func(x, a, b, c, d,e,f):
    x1 = x[:, 0]  # 提取x1数据
    x2 = x[:, 1]  # 提取x2数据
    return a * ((x1 / (x1 + 1)) * ((x2 + f) / (x2 + 0.2)) * ((x2 + b * x1 + c) / (x2 + d * x1 + e)))

# 提取数据
x1_data = df.iloc[:, 5].values
x2_data = df.iloc[:, 6].values
y_data = df.iloc[:, 0].values

# 创建参数数组
params_array = np.column_stack((x1_data, x2_data))

# 使用curve_fit进行非线性回归拟合
popt, pcov = curve_fit(nonlinear_func, params_array, y_data, p0=popt0,maxfev=1000000)
# 或者使用 least_squares 函数
from scipy.optimize import least_squares

initial_params = np.array(popt0)
#result = least_squares(nonlinear_func, initial_params, args=(params_array, y_data))

#popt_least_squares = result.x
#pcov_least_squares = result.covar
# 使用拟合参数进行预测
y_pred = nonlinear_func(params_array, *popt)

# 计算R²分数
r2 = r2_score(y_data, y_pred)
mse=mean_squared_error(df.iloc[:, 0], y_pred)
mae=mean_absolute_error(df.iloc[:, 0], y_pred)
# 打印拟合参数和R²值
print("拟合参数：a={}, b={}, c={}, d={},e={},f={}".format(*popt))
print("R²值：", r2)
print("MAE值：", mae)
print("MSE值：", mse)
```
